#### loc [ ] vs iloc [ ]

```python
# ✅ loc[] – Label-based indexing

import pandas as pd

df = pd.DataFrame({

     'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['NY', 'LA', 'Chicago']
    
}, index=['a', 'b', 'c'])

print(df)
    
```

```python
# Get row with label 'a'

print(df.loc['a'])


# Get rows 'a' to 'b' (inclusive!)

print(df.loc['a':'b'])

# Get specific value: Age of person in row 'a'

print(df.loc['a', 'Age'])

# Get all rows, only 'Name' column

print(df.loc[:, 'Name'])

# Get specific rows and columns

print(df.loc[['a', 'c'],['Name','City']])

# Use boolean filtering with loc

print(df.loc[df['Age'] > 28 ])
```

```python
# ✅ iloc[] – Integer-location-based indexing

import pandas as pd

# Sample DataFrame
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['NY', 'LA', 'Chicago']
})

print(df)
```

```python
# Get the first row

print(df.iloc[0])

# Get first two rows

print(df.iloc[0:2])

# Get first row, second column (Age of Alice)

print(df.iloc[0,1])

# Get all rows, only the first column

print(df.iloc[:,0])

# Get last row using negative index

print(df.iloc[-1])

# Get specific rows and columns using list of indices

print(df.iloc[[0,2],[1,2]])
```

#### Duplicates

```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Alice', 'David'],
    'Age': [25, 30, 25, 40]
})

print(df)
```

```python
# Find duplicate rows (returns a Boolean Series)

print(df.duplicated())

# Get only duplicated rows

print(df[df.duplicated()])

# Based on a specific column

print(df[df.duplicated(subset = 'Name')])

# Mark all duplicates as True (including the first occurrence)

print(df[df.duplicated(keep=False)])
```

```python
# 🔁 Combined Workflow Example:

# Step 1: Find duplicates
print(df[df.duplicated(subset='Name')])

# Step 2: Remove duplicates
df_clean = df.drop_duplicates(subset='Name')

print(df_clean)
```

```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Alice', 'David'],
    'Age': [25, 30, 25, 40]
})

print(df)
```

```python
# Drop completely duplicate rows

df_unique = df.drop_duplicates()
print(df_unique)

# Drop duplicates based on one column

print(df.drop_duplicates(subset = 'Name'))

# Keep the last duplicate

print(df.drop_duplicates(subset = 'Name', keep = 'last'))

# Drop all duplicates (keep none)

print(df.drop_duplicates(subset = 'Name', keep = False))


```

#### Combine Comumns

You can combine columns in Pandas in a few different ways, depending on what you're trying to do:

#### Group By

```python
import pandas as pd

df = pd.DataFrame({
    'Department': ['Sales', 'Sales', 'HR', 'HR', 'IT'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Salary': [50000, 55000, 60000, 62000, 70000]
})

print(df)
```

```python
# Group by Department and get an average salary

df.groupby('Department')['Salary'].mean()
```

```python
# Multiple Aggregations (with agg()):

df.groupby('Department').agg({
    'Salary': ['mean', 'max', 'min']
})
```

```python
# Group by Multiple Columns:

df.groupby(['Department','Employee'])['Salary'].sum()
```

```python
# Filter Groups: Show only departments with total salary > 100000

df.groupby('Department').filter(lambda x : x['Salary'].sum() > 100000).sort_values(by = 'Department', ascending = True)
```

#### Merge

```python
import pandas as pd

df1 = pd.DataFrame({
    'EmployeeID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})

df2 = pd.DataFrame({
    'EmployeeID': [2, 3, 4],
    'Department': ['HR', 'IT', 'Sales']
})


merged_df = pd.merge(df1, df2, on='EmployeeID', how='inner')

print(merged_df)
```

```python
# Custom Column Keys:

pd.merge(df1, df2, left_on =1, right_on =2)
```

```python
# Merge on Index:

pd.merge(df1, df2, left_index=True, right_index=True)
```

```python
# Handling Overlapping Columns:

pd.merge(df1, df2, on='EmployeeID', suffixes=('_left', '_right'))
```

#### Assign

```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob'],
    'Marks': [80, 90]
})

df = df.assign(Grade = lambda x: ['A' if m > 85 else 'B' for m in x['Marks']])

print(df)
```

```python
# You can create multiple columns at once:

df = df.assign(
    Bonus = lambda x: x['Marks'] * 0.1,
    Final = lambda x: x['Marks'] + x['Marks'] * 0.1
)

print(df)
```

```python
# 🔁 Comparison:

# Traditional
df['Bonus'] = df['Marks'] * 0.1

print(df['Bonus'])

# With assign
df = df.assign(Bonus = lambda x: x['Marks'] * 0.1)

print(df)
```

```python
# Example using groupby(), filter(), and .assign() — great for a real-world data processing pipeline.

# Scenario: Analyze employee salaries per department

import pandas as pd

# Sample data
df = pd.DataFrame({
    'Department': ['Sales', 'Sales', 'HR', 'HR', 'IT', 'IT'],
    'Employee': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Salary': [50_000, 55_000, 60_000, 62_000, 70_000, 75_000]
})

# Step 1: Group by department

grouped = df.groupby('Department')


# Step 2: Filter departments with total salary > 110000

filtered = grouped.filter(lambda x: x['Salary'].sum() > 110000)


# Step 3: Assign extra columns (e.g., bonus, final salary)

result = (
    filtered
    .assign(
        Bonus=lambda x: x['Salary'] * 0.10,
        FinalSalary=lambda x: x['Salary'] + x['Salary'] * 0.10
    )
)


print(result)
```

#### PIVOT Table

```python
import pandas as pd

df = pd.DataFrame({
    'Department': ['Sales', 'Sales', 'HR', 'HR', 'IT', 'IT'],
    'Gender': ['M', 'F', 'F', 'M', 'M', 'F'],
    'Salary': [50000, 55000, 60000, 62000, 70000, 72000]
})

print(df)
```

```python
# Create a pivot table: Average salary by Department and Gender

pd.pivot_table(
    df,
    index='Department',
    columns='Gender',
    values='Salary',
    aggfunc='mean'
)
```

```python
#  Add Totals with margins=True:

pd.pivot_table(
    df,
    index='Department',
    columns='Gender',
    values='Salary',
    aggfunc='mean',
    margins=True,
    margins_name='Total'
)
```

```python
# Multiple Aggregations with aggfunc:

pd.pivot_table(
    df,
    index='Department',
    values='Salary',
    aggfunc=['mean', 'max', 'count']
)
```

```python
# Fill Missing Values:

pd.pivot_table(
    df,
    index='Department',
    columns='Gender',
    values='Salary',
    aggfunc='mean',
    fill_value=0
)
```

```python
# If you want to reshape pivoted data back, use:

df.pivot_table(...).reset_index()
```

#### Timeseries

📈 Time Series in Pandas

Time series data is data indexed by timestamps — Pandas makes working with dates & times fast and flexible.

```python
# 1. Datetime Index Basics

# Convert a column to datetime:

df['Date'] = pd.to_datetime(df['Date'])

# Set it as index:

df.set_index('Date', inplace=True)
```

```python
# 2. Resampling (Group by Time) : Daily to Monthly Average

df.resample('M').mean()
```

```python
# 3. Date Filtering

df['2023']              # all data in 2023
df['2023-05']           # May 2023
df['2023-05-15']        # Only that day
```

```python
# 4. Date Features

df['Year'] = df.index.year
df['Month'] = df.index.month
df['Day'] = df.index.day_name()
```

```python
# 5. Rolling Window Calculations

df['RollingAvg'] = df['Value'].rolling(window=7).mean()
```

```python
# 6. Shift / Lagging Data

df['Previous'] = df['Value'].shift(1)
```

```python
# 🧪 Example:

df = pd.DataFrame({
    'Date': pd.date_range(start='2023-01-01', periods=10, freq='D'),
    'Sales': [100, 120, 90, 130, 115, 150, 160, 170, 180, 190]
})
df.set_index('Date', inplace=True)

monthly_avg = df.resample('ME').mean()

print(monthly_avg)
```
